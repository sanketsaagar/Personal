```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('averaged_perceptron_tagger')
```

```python
import pandas as pd
from apyori import apriori
```

```python
data = pd.read_csv('aisles.csv')
data.head(5)
data.shape
```

```python
a=[]
for i in data['aisle']:
    text = word_tokenize(i)
    a.append(nltk.pos_tag(text))
```

```python
a
```

```python
out_tup = [i for i in a[0] if i[1] == 'NNS']
out_tup
```

```python
updated_list =[]
for j in a:
    for i in j:
        if(i[1] == 'NNS' or i[1] == 'NN'):
            updated_list.append(i)
```

```python
print(updated_list)
```

```python
df = pd.DataFrame(updated_list)
df
```

```python
df.shape
```

```python
records = []
for i in range(0,df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(0,df.shape[1])])
records
```

```python
rules = apriori(records,min_support = 0.01,min_confidence = 0.6, min_lift = 1.2,min_length=2)
result = list(rules)
print(result)
```
